```python
## Ordinal features
ordinal={
'Type of Admission':['Emergency','Urgent','Trauma'],
'Severity of Illness':['Minor','Moderate','Extreme'],
'Stay':['0-10','11-20','21-30','31-40','41-50','51-60','61-70','71-80','81-90','91-100','More than 100 Days']
}

## Norminal features
norminal_feat ={
'Hospital_type_code':'hos_code',
'Hospital_region_code':'hos_reg',
'Department':'dep',
'Ward_Type':'ward_type',
'Ward_Facility_Code':'ward_facility',
'Age':'age'
}
```

```python
def binary_encode(data, col, pos_val):
    data[col]=data[col].apply(lambda x: 1 if x==pos_val else 0)
    return data

def ordinal_encode(data, ordinal):
    for col,order in ordinal.items():
        data[col]=data[col].apply(lambda x: order.index(x))
    return data

def norminal_encode(data, norminal_feat):
    for col,title in norminal_feat.items():
        dummies=pd.get_dummies(data[col],prefix=title)
        data=pd.concat([data, dummies], axis=1)
        data=data.drop(col,axis=1)
    
    return data
```

```python
df=norminal_encode(df, norminal_feat)
```

```python
df=ordinal_encode(df, ordinal)
```

```python
import xgboost
from sklearn.model_selection import train_test_split
```

```python
from sklearn.ensemble import AdaBoostClassifier,RandomForestClassifier
from xgboost import XGBClassifier,XGBRFClassifier
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=50)

xtrain.shape,xtest.shape,ytrain.shape,ytest.shape
```

```python
from collections import Counter
```

```python
Counter(ytrain)
```

```python
models={
AdaBoostClassifier():'AdaBoostClassifier',
RandomForestClassifier():'RandomForestClassifier',
XGBClassifier():'XGBClassifier',
XGBRFClassifier():'XGBRFClassifier'
}
```

```python
for model, name in models.items():
    model.fit(xtrain,ytrain)
```

```python
for model, name in models.items():
    print(name, accuracy_score(ytest, model.predict(xtest))*100)
```
